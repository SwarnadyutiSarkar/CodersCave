```python
X = df['text']
Y= df ['spam']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
X = cv.fit_transform(X) # Fit the Data
```

```python

feature_names = cv.get_feature_names_out()


df = pd.DataFrame(X.toarray(), columns=feature_names)


df.head()
```

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB


clf = MultinomialNB()
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score 

y_pred = clf.predict(x_test)
accuracy = accuracy_score(y_pred,y_test)

print("Accuracy:", accuracy )
```

```python

new_email = "This is a promotional email. Please click the link to get a free discount!"
new_email_cv = cv.transform([new_email])
prediction = clf.predict(new_email_cv)


if prediction == 0:
    print("New email is classified as ham")
else:
    print("New email is classified as spam")
```
